```python
import matplotlib.pyplot as plt
import mplcursors

number = int(input("Enter a positive integer: "))
```

```python
sequence = [number]
while number != 1:
    itrs = itrs +1
    if number % 2 == 0:
        number = number // 2
    else:
        number = number * 3 + 1
    sequence.append(number)
```

```python
print("Sequence: ",sequence)
print("Total iterations :", len(sequence)-1)
```

```python
plt.plot(sequence)
plt.xlabel('Iteration')
plt.ylabel('Number')
plt.title('Collatz Conjecture')
plt.grid(True)

# Find the crest and its index in the sequence
crest_value = max(sequence)
crest_index = sequence.index(crest_value)

# Mark the crest with its value
plt.annotate(f'Crest: {crest_value}', xy=(crest_index, crest_value), xytext=(crest_index, crest_value + 10),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.show()
```
